```python
import pandas as pd
import json

# only extract the columns we need from the mega table: jobs
jobs_df = mega_table[['name','job_id', 'title', 'description_x', 'years_of_experience', 'detailed_description', 
                      'responsibilities', 'requirements', 'average_salary', 'location', 'industry_name']]

# Define the bins and labels for the experience levels
bins = [-1, 3, 7, 10]
labels = ['Entry Level', 'Mid Level', 'Senior Level']

# Use pd.cut() to create the 'experience_level' column
jobs_df['experience_level'] = pd.cut(jobs_df['years_of_experience'], bins=bins, labels=labels)

# Display the updated DataFrame
print(jobs_df)


jobs = {}
jobs['jobs'] = jobs_df.to_dict(orient='records')

# Save transformed data as JSON file(s)
with open('jobs.json', 'w') as f:
    json.dump(jobs, f)


# only extract the columns we need from the mega table: companies
companies_df = mega_table[['name','size','type','location','website','description_x','hr_contact']]
companies = {}
companies['companies'] = companies_df.to_dict(orient='records')

with open('companies.json', 'w') as f:
    json.dump(companies, f)


# only extract the columns we need from the mega table: industries
industries_df = mega_table[["industry_name", "growth_rate", "industry_skills", "top_companies", "trends"]]
industries = {}
industries['industries'] = industries_df.to_dict(orient='records')

with open('industries.json', 'w') as f:
    json.dump(industries, f)


```

```python
import csv 
import json
import pymongo

# connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["careerhub"]


# define collections
jobs = db["jobs"] 
companies = db["companies"]
industries = db["industries"]

# loop through industries.json rows and insert into MongoDB
with open('industries.json', 'r') as file:
  data = json.load(file)
  for row in data['industries']:
    print(row)
    industries.insert_one(row)

# loop through companies.json rows and insert into MongoDB
with open('companies.json', 'r') as file:
  data = json.load(file)
  for row in data['companies']:
    print(row)
    companies.insert_one(row)


# loop through jobs.json rows and insert into MongoDB
with open('jobs.json', 'r') as file:
  data = json.load(file)
  for row in data['jobs']:
    print(row)
    jobs.insert_one(row)
  
```
